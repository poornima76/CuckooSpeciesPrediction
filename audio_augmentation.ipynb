```python
#1
#Splitting audio files

import os
import pydub
from pydub import AudioSegment
pydub.AudioSegment.ffmpeg = "/absolute/path/to/ffmpeg"
# Define the input folders(import downloaded audio files)
input_dir = "downloads\Indian_Cuckoo2"
# Define the output folders(exoprt split audio files)
output_dir = "audio\Indian_Cuckoo"

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through all the audio files in the input directory
for file_name in os.listdir(input_dir):
    if file_name.endswith(".mp3") or file_name.endswith(".wav"):
        print(os.path.join(input_dir, file_name))
        # Load the audio file
        audio = AudioSegment.from_file(os.path.join(input_dir, file_name))
        # Split the audio file into 10-second chunks
        chunks = audio[::10000]

        # Save each chunk to the output directory with a numbered file_name
        for i, chunk in enumerate(chunks):
            chunk.export(os.path.join(output_dir, f"{os.path.splitext(file_name)[0]}_{i+18:03}.wav"), format="wav")
```

```python
#2
#deleting audio files that are shorter than 10s
import os
import pydub
from pydub import AudioSegment
def delete_short_audio_files(folder, min_duration=10000):
    # Process each audio file in the folder
    for file_name in os.listdir(folder):
        if file_name.endswith(".mp3") or file_name.endswith(".wav"):
            file_path = os.path.join(folder, file_name)
            audio = AudioSegment.from_mp3(file_path)

            # Check if the audio duration is less than the minimum
            if len(audio) < min_duration:
                os.remove(file_path)
                print(f"Deleted short audio file: {file_name}")

#path of input folder(folder with split audio files) 

input_folder = "audio\Indian_Cuckoo"

# Delete short audio files
delete_short_audio_files(input_folder)
```

```python
#checking length of each audio files
import os
import pydub
from pydub import AudioSegment
def check_audio_files(folder, min_duration=10000):
    # Process each audio file in the folder
    count=0
    count2=0
    for file_name in os.listdir(folder):

        if file_name.endswith('.mp3') or file_name.endswith(".wav"):
            file_path = os.path.join(folder, file_name)
            audio = AudioSegment.from_mp3(file_path)
            # Check if the audio duration is less than the minimum
            if len(audio)> min_duration:
                count=count+1
            if len(audio) == min_duration:
                count2=count2+1

    print(count) 
    print(count2)


#path of input folder(folder with split audio files)

input_folder = "audio\Indian_Cuckoo"

# Delete short audio files
check_audio_files(input_folder)
```

```python

#converting mp3 files to wav files
from pydub import AudioSegment
import os

def convert_folder_mp3_to_wav(input_folder):
    # Process each file in the input folder
    for file_name in os.listdir(input_folder):
        if file_name.endswith('.mp3'):
            mp3_file_path = os.path.join(input_folder, file_name)

            # Load the MP3 file and convert to WAV
            audio = AudioSegment.from_mp3(mp3_file_path)

            # Define the WAV file path (replace extension)
            wav_file_path = os.path.join(input_folder, os.path.splitext(file_name)[0] + ".wav")

            # Export the audio to WAV format
            audio.export(wav_file_path, format="wav")

            print(f"Conversion complete for: {file_name}. WAV file replaced at: {wav_file_path}")

            # Remove the original MP3 file
            os.remove(mp3_file_path)

# Path of input folder -new files will replace the original mp3 file and be saved in the same folder
input_folder = "pacific_ten"
convert_folder_mp3_to_wav(input_folder)
# file_name = "Pacific_Koel034.mp3" 
# if file_name.endswith('.mp3'):
#         mp3_file_path = os.path.join(input_folder, file_name)

#             # Load the MP3 file and convert to WAV
#         audio = AudioSegment.from_mp3(mp3_file_path)
#         wav_file_path = os.path.join(input_folder, os.path.splitext(file_name)[0] + ".wav")

#             # Export the audio to WAV format
#         audio.export(wav_file_path, format="wav")
#         os.remove(mp3_file_path)
```

```python
#renaming files into proper names

import os

def rename_files(folder_path, prefix):
    # List all files in the folder
    files = os.listdir(folder_path)

    # Sort files to ensure consistent numbering
    files.sort()

    # Iterate through each file and rename
    for i, file_name in enumerate(files):
        file_extension = os.path.splitext(file_name)[1]
        new_name = f"{prefix}{i+1:03d}{file_extension}"
        old_path = os.path.join(folder_path, file_name)
        new_path = os.path.join(folder_path, new_name)

        # Rename the file
        os.rename(old_path, new_path)

        print(f"Renamed: {file_name} to {new_name}")

#Path of input folder
folder_path = "audio\Indian_Cuckoo"
#name of the audio file
prefix = "Indian_Cuckoo"

# Rename files in the folder
rename_files(folder_path, prefix)
```
